## Step 2: Clean your data.

##### Do some rudimentary cleaning. Rename any columns that are poorly named, shorten any strings that may be too long, fill in any missing values. Explain your rationale for the way you choose to "impute" the missing data.

```python


df = df.rename(columns={'artist.inverted':'Artist', 'time':'Length', 'date.entered':'Entered', 'date.peaked':'Peak'})

#Clean names of week columns
df.columns = df.columns.str.replace('.week','')
df.columns = df.columns.str.replace('x','')
df['1st'] = df['1st'].astype(float) 

df.columns
```

```python
#df2.dtypes
import re

def removeAsciiCharacters(x): return re.sub(r'[^\x00-\x7f]',r'',x)

df["Artist"] = df.Artist.apply(removeAsciiCharacters)
#df2['Rank in Top 100'] = df2['Rank in Top 100'].apply(removeAsciCharacters)


df
```

##### Using Pandas' built in `melt` function, pivot the weekly ranking data to be long rather than wide. As a result, you will have removed the 72 'week' columns and replace it with two: Week and Ranking. There will now be multiple entries for each song, one for each week on the Billboard rankings.

```python
df1 = pd.melt(df, id_vars=['year', 'Artist', 'track','Length','genre','Entered','Peak'], var_name='Week',
              value_name = 'Rank')

df1['Entered']= pd.to_datetime(df1['Entered'])
df1['Peak']= pd.to_datetime(df1['Peak'])
df1['Days to Peak'] = (df1['Peak']-df1['Entered']).dt.days
df1['Days to Peak'] = df1['Days to Peak']
df1.dropna()
```

```python
df1['Week'] = df1['Week'].str.replace('st','')
df1['Week'] = df1['Week'].str.replace('rd','')
df1['Week'] = df1['Week'].str.replace('nd','')
df1['Week'] = df1['Week'].str.replace('th','')
df1['Week'] = df1['Week'].astype(int)
df1.dtypes
```

```python
df1.dtypes
```

## Step 3: Visualize your data.

```python
df2 = df1['Days to Peak'].groupby(df1['genre']).mean()

df2df = pd.DataFrame(df2)
df2df['genre'] = df2.index
df2.plot.bar(x="genre", y="Days to Peak")
plt.ylabel('Average Days to Peak Rank in Top 100')
plt.xlabel('Genre')
df2df
```

```python
df3 = df1['Rank'].groupby(df1['genre']).count()
df3
#df1.dtypes
df3rf = pd.DataFrame(df3)
df3rf ['genre'] = df3rf.index
df3rf ['Times Ranked in top 100'] = df3rf[0]

df3rf

df3rf.plot.barh(x=['genre'], y=['Times Ranked in top 100'])
plt.xlabel('Times Ranked in top 100')
plt.ylabel('Genre')
```
