##### 19. Plot some scatterplots. **BONUS**: Use a PyPlot `figure` to present multiple plots at once.

```python
sns.jointplot(db['Verbal'], db['Math'], kind='reg', size=10, joint_kws={'scatter_kws': {'s': 150}})
plt.title("SAT Verbal vs. Math Scatterplot")
plt.xlabel("Verbal")
plt.ylabel("Math")
plt.show()

sns.jointplot(db['Verbal'], db['Rate'], kind='reg', size=10, joint_kws={'scatter_kws': {'s': 150}})
plt.title("SAT Verbal vs. Participation Rate Scatterplot")
plt.xlabel("Verbal")
plt.ylabel("Rate")
plt.show()

sns.jointplot(db['Math'], db['Rate'], kind='reg', size=10, joint_kws={'scatter_kws': {'s': 150}})
plt.title("SAT Math vs. Participation Rate Scatterplot")
plt.xlabel("Verbal")
plt.ylabel("Math")
plt.show()
```
